```python
import pandas as pd
import numpy as np
import chardet
```

```python
with open('database_cleaned.csv', 'rb') as f:
    result = chardet.detect(f.read(10000))
    encoding = result['encoding']
    
print(f'Codificación detectada: {encoding}')
```

```python
# Leer el archivo CSV con la codificación detectada
df = pd.read_csv('database_cleaned.csv', encoding='ISO-8859-1', delimiter=',')
```
